```python
import csv

def main():
    f = open('q4.csv','r',encoding='cp949')
    data = csv.reader(f)
    header = next(data)
    sation_max = ['','','','']
    sation_low= ['','','','']

    max_total = [0,0,0,0]
    low_total = [float('inf'),float('inf'),float('inf'),float('inf')]

    for row in data:
        if(row[1]== "1호선"):
            idx = 0
            row[-2] = int(row[-2])
            row[-3] = int(row[-3])
            total = row[-2]+row[-3]
            if(total>max_total[idx]):
                sation_max[idx] = row[3]
                max_total[idx] = total
            if(total<low_total[0]):
                sation_low[idx] = row[3]
                low_total[0] = total
        elif(row[1]=="2호선"):
            idx = 1
            row[-2] = int(row[-2])
            row[-3] = int(row[-3])
            total = row[-2]+row[-3]
            if(total>max_total[idx]):
                sation_max[idx] = row[3]
                max_total[idx] = total
            if(total<low_total[idx]):
                sation_low[idx] = row[3]
                low_total[idx] = total
        elif(row[1]=="3호선"):
            idx = 2
            row[-2] = int(row[-2])
            row[-3] = int(row[-3])
            total = row[-2]+row[-3]
            if(total>max_total[idx]):
                sation_max[idx] = row[3]
                max_total[idx] = total
            if(total<low_total[idx]):
                sation_low[idx] = row[3]
                low_total[idx] = total
        elif(row[1]=="4호선"):
            idx = 3
            row[-2] = int(row[-2])
            row[-3] = int(row[-3])
            total = row[-2]+row[-3]
            if(total>max_total[idx]):
                sation_max[idx] = row[3]
                max_total[idx] = total
            if(total<low_total[idx]):
                sation_low[idx] = row[3]
                low_total[idx] = total
        
        
    
    f.close()

    print("*** Subway Report for Seoul on March 2023 ***")
    for i in range(0,4):
        print("LIne ",i+1,":")
        print("Busiest Station:",sation_max[i], max_total[i])
        print("Least uese Station:",sation_low[i], low_total[i])
        
    
if __name__ == "__main__":
    main()
```
