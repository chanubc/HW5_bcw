```python
import csv

def main():
    f = open('Seoul_Temp.csv','r',encoding='cp949')
    data = csv.reader(f)
    header = next(data)
    #print(header)
    max_temp = 0
    low_temp =0
    avr_temp = 0
    avr_low = 0
    avr_max =0
    count = 0

    for row in data:
        if(row[-1]==''):
            row[-1] = 0
        row[-1] = float(row[-1])
        if max_temp < row[-1]:
            max_date = row[0]
            max_temp = row[-1]
        
        if(row[-2]==''):
            row[-2] = 0
        row[-2] = float(row[-2])
        if low_temp > row[-2]:
            low_date = row[0]
            low_temp = row[-2]
            
        #연평균 기온
        count = count+1
        if(row[2]==''):
            row[2] = 0
        row[2] = float(row[2])
        avr_temp = row[2]+avr_temp
        
        #연평균 최저 기온
        if(row[-2]==''):
            row[-2]=0
        row[-2] = float(row[-2])
        avr_low = row[-2]+avr_low
        
        #연평균 최고 기온
        if(row[-1]==''):
            row[-1]=0
        row[-1] = float(row[-1])
        avr_max = row[-1]+avr_max


    avr_temp = avr_temp/count
    avr_low = avr_low/count
    avr_max = avr_max/count
    
    avr = format(avr_temp,".2f")
    avr_low = format(avr_low,".2f")
    avr_max = format(avr_max,".2f")
    
    f.close()
    print("*** Anunual Temperature Report for Seoul in 2022 ***")
    print("Average Temperature:",avr,"Celsius")
    print("Average Minimum Temperature:",avr_low,"Celsius")
    print("Average Maximum Temperature:",avr_max,"Celsius")
        
    
if __name__ == "__main__":
    main()
```

```python
import csv

def main():
    f = open('Seoul_Subway.csv','r',encoding='cp949')
    data = csv.reader(f)
    header = next(data)
    sation_max = ['','','','']
    sation_low= ['','','','']

    max_total = [0,0,0,0]
    low_total = [float('inf'),float('inf'),float('inf'),float('inf')]

    for row in data:
        if(row[1]== "1호선"):
            idx = 0
            row[-2] = int(row[-2])
            row[-3] = int(row[-3])
            total = row[-2]+row[-3]
            if(total>max_total[idx]):
                sation_max[idx] = row[3]
                max_total[idx] = total
            if(total<low_total[0]):
                sation_low[idx] = row[3]
                low_total[0] = total
        elif(row[1]=="2호선"):
            idx = 1
            row[-2] = int(row[-2])
            row[-3] = int(row[-3])
            total = row[-2]+row[-3]
            if(total>max_total[idx]):
                sation_max[idx] = row[3]
                max_total[idx] = total
            if(total<low_total[idx]):
                sation_low[idx] = row[3]
                low_total[idx] = total
        elif(row[1]=="3호선"):
            idx = 2
            row[-2] = int(row[-2])
            row[-3] = int(row[-3])
            total = row[-2]+row[-3]
            if(total>max_total[idx]):
                sation_max[idx] = row[3]
                max_total[idx] = total
            if(total<low_total[idx]):
                sation_low[idx] = row[3]
                low_total[idx] = total
        elif(row[1]=="4호선"):
            idx = 3
            row[-2] = int(row[-2])
            row[-3] = int(row[-3])
            total = row[-2]+row[-3]
            if(total>max_total[idx]):
                sation_max[idx] = row[3]
                max_total[idx] = total
            if(total<low_total[idx]):
                sation_low[idx] = row[3]
                low_total[idx] = total
        
        
    
    f.close()

    print("*** Subway Report for Seoul on March 2023 ***")
    for i in range(0,4):
        print("LIne ",i+1,":")
        print("Busiest Station:",sation_max[i], max_total[i])
        print("Least uese Station:",sation_low[i], low_total[i])
        
    
if __name__ == "__main__":
    main()
```
